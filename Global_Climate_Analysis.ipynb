# Global Temp

Here, we will overlay all of the datasets previously imported together on a matplotlib.pyplot figure using line graphs.

```python
plt.plot(climate_1['Year'], climate_1['Temp.Anom'], label='0 N to 30 N')
plt.plot(climate_1['Year'], climate_2['Temp.Anom'], label='0 N to 90 N')
plt.plot(climate_1['Year'], climate_3['Temp.Anom'], label='20 N to 90 N')
plt.plot(climate_1['Year'], climate_4['Temp.Anom'], label = '20 S to 20 N')
plt.plot(climate_1['Year'], climate_5['Temp.Anom'], label = '30 N to 60 N')
plt.plot(climate_1['Year'], climate_6['Temp.Anom'], label = '30 S to 00 N')

plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (K)')
plt.title('Global Temperature Analysis')
plt.legend()
plt.grid(True)
```

From this plot, we can get a sense of how the data is distributed. We also are given the chance to see if any of the datasets obviously differ from the rest.

We can see that, for the most part, all of the datasets follow the same trend. This corroborates the idea that you cannot see a change in temperature in some part of the world without seeing at least a small version of that same change in another part of the world.

```python
# Create a dataframe in pandas consisting of all the anomalies from each dataset

Anomalies = pd.DataFrame(data =  {'C_1': climate_1.iloc[:,1],
                                  'C_2': climate_2.iloc[:,1],
                                  'C_3': climate_3.iloc[:,1],
                                  'C_4': climate_4.iloc[:,1],
                                  'C_5': climate_5.iloc[:,1],
                                  'C_6': climate_6.iloc[:,1]})
```

```python
Anomalies.head()
```

Now, we can perform Principal Component Analysis (PCA) to determine which latitudinal range has the biggest effect on the global temperature.

```python
# Normalize data before PCA.
# We will be using MaxAbsScaler to normalize the data due to the number of
#   negative values seen in the data.

### Normalize
normalized_anom = pd.DataFrame(data = MaxAbsScaler().fit_transform(Anomalies), columns = Anomalies.columns)

### Do PCA
pca = PCA(n_components = normalized_anom.shape[1])
pcaData = pca.fit_transform(normalized_anom)

### Extract Components
PC = pca.components_

### Create DataFrame of components to understand them
PC_df = pd.DataFrame(data = PC, index = normalized_anom.columns, columns = ('PC' + str(i) for i in range(normalized_anom.shape[1])))
print(PC_df.iloc[:, 0:4])
```

Now that we have our PCA components, we will create a PCA Biplot in order to visualize the results, allowing us to see which latitudinal range of anomalies has the biggest impact on the global environment.

```python
### Create Figure
fig = plt.figure(figsize = (20, 10))

### Plot data points in new PCA space
for i in range(len(pcaData)):
  plt.scatter(pcaData[i,0], pcaData[i,1], marker = 'o', c = 'white', alpha = 0.5)

### Plot arrows with labels to see which latitudinal range has biggest impact
scale = 10
for k in range(normalized_anom.shape[1]):
  plt.text(PC_df.iloc[k,0] * scale + 0.05,
           PC_df.iloc[k,1] * scale + 0.05,
           PC_df.index[k],
           fontweight = 'bold', fontsize = 12)
  plt.arrow(0, 0, PC_df.iloc[k, 0] * scale, PC_df.iloc[k, 1] * scale, color = 'r', head_width = 0.15, linewidth = 0.5, alpha = 0.5)

### Label the Biplot axes and title
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Temperature Anomalies PCA Biplot')
```

In a PCA Biplot, the arrows represent the amount of impact each dataset's primary principal component has on the variance in the data.

From this biplot, we can see that the biggest impact is from C_3 due to the length of the arrow, indicating the largest impact.

Looking at our graph, we can see that the USA1 location has the most annual plastics waste per year.

Now, we will further our analysis using the USA1 plastics data.

```python
### Check the earliest year in the plastics waste data

min(plastics['TIME_PERIOD'])
```

```python
### Check the most recent year in the plastics waste data

max(plastics['TIME_PERIOD'])
```

```python
### Check the earliest year in the climate_3 data

min(climate_3['Year'])
```

```python
### Check the most recent year in the climate_3 data

max(climate_3['Year'])
```

```python
### Filter the climate_3 data to years matching the time range on the plastics data

climate_3_1990_2019 = climate_3[np.logical_and(climate_3['Year'] >= 1990, climate_3['Year'] <= 2019)]
climate_US_plastics = climate_3_1990_2019.copy()

### Add the plastics waste data to the filtered climate_3 dataframe to run
###   correlation analysis

climate_US_plastics['Plastic.Waste'] = plastics_srted[plastics_srted['LOCATION'] == 'USA1']['OBS_VALUE'].values
climate_US_plastics.head()
```

```python
### Check the correlation coefficients

correlations_plastics = climate_US_plastics.corr()
correlations_plastics.head()
```

Looking at the correlation coefficients, we can see that Plastic.Waste and Temp.Anom have a correlation coefficient of 0.830074

Let's determine if the correlation coefficient of 0.830074 between Plastic.Waste and Temp.Anom is significant.

We will calculate the t-value using the equation:

$$t = \frac{r * \sqrt{n - 2}}{\sqrt{1 - r^2}}$$
Where:
- r is the correlation coefficient
- n is the number of observations

```python
### Define our r value
r = correlations_plastics['Plastic.Waste'][1]

### Define n value
n = len(climate_US_plastics)

### Define t value
t = (r * np.sqrt(n - 2)) / np.sqrt(1 - r ** 2)
```

```python
print("t-value: ", t)
print("\nCompare to t-dist table with degrees of freedom = (n - 2): 28")
print("\nFor two-tailed significance value of 0.05, critical t-value: 2.048\n")
print(t, "is far greater than 2.048, thus, our result is statistically significant!")
```

Now, we can conclude that the plastics waste of the United States has a statistically significant correlation value in association with the temperature anomalies seen from 1990 to 2019 in the latitudinal range 20 N to 90 N.

We also know that correlation does not equal causation. Thus, we must continue to work towards proving causation.

# Temporal Precedence

Let us work towards causation by establishing temporal precedence.
- **Temporal Precedence** - the principle that for one variable, cause, to be considered as influencing another variable, effect, the cause must occur before the effect in time

```python
### Create a new dataframe that is a copy of climate_US_oil
climate_US_oil_2 = climate_US_oil.copy()
climate_US_oil_2.drop('Year', axis = 1, inplace = True) # Drop year
US_oil_std = MaxAbsScaler().fit_transform(climate_US_oil) # Standardize data
# Create normalized dataframe
US_oil_std_df = pd.DataFrame(data = US_oil_std, columns = climate_US_oil.columns, index = climate_US_oil.index)
```

```python
### Create plot, overlaying Temp.Anom and Oil Production
plt.plot(climate_US_oil['Year'], US_oil_std_df['Temp.Anom'])
plt.plot(climate_US_oil['Year'], US_oil_std_df['Oil Production (TWh)'])
plt.plot(2008, 0.4, marker = 'o', color = 'r', markersize = 20, alpha = 0.5)
plt.plot(2012, 0.5, marker = 'o', color = 'y', markersize = 20, alpha = 0.5)
plt.legend(labels = ['Temp.Anom', 'Oil Production (TWh)'])
plt.title('US Oil Production and Temperature Anomalies Under MaxAbsScaler')
```

Above, we can see the temporal precedence, showing that the trend in oil production each year can also be seen in the future temperature anomalies.

The red and yellow dots showcase a clear precedence as the oil production showed a decrease, and, in years following, we see a decrease in temp.anom.

This is seen throughout the entirety of the graph.

Now, we will look for temporal precedence in plastics waste.

```python
### Create a new dataframe that is a copy of climate_US_plastics
climate_US_plastics_2 = climate_US_plastics.copy()
climate_US_plastics_2.drop('Year', axis = 1, inplace = True) # Drop Year
US_plastics_std = MaxAbsScaler().fit_transform(climate_US_plastics_2) # Standardize data
# Create normalized dataframe
US_plastics_std_df = pd.DataFrame(data = US_plastics_std, columns = climate_US_plastics_2.columns, index = climate_US_plastics_2.index)
```

```python
### Create plot, overlaying Temp.Anom and Plastic.Waste
plt.plot(climate_US_plastics['Year'], US_plastics_std_df['Temp.Anom'])
plt.plot(climate_US_plastics['Year'], US_plastics_std_df['Plastic.Waste'])
plt.legend(labels = ['Temp.Anom', 'Plastic.Waste'])
plt.title('US Plastic Waste and Temperature Anomalies Under MaxAbsScaler')
```

It is harder to see the change in the plastics waste graph, thus making it harder to see if there is any temporal precedence in the plastics waste and temperature anomalies graph, so we will try to visualize it better in our upcoming linear regression models.

# Linear Regression Model

Let's first take a further look into the oil production using a linear regression model from the sklearn package.

```python
from sklearn.linear_model import LinearRegression
```

```python
# Oil Production LRM
### Create and fit linear regression model to the standardized oil data
lrm = LinearRegression()
lrm.fit(US_oil_std_df['Oil Production (TWh)'].values.reshape(-1, 1), US_oil_std_df['Temp.Anom'].values)

### Make our prediction (regression) line
temp_preds = lrm.predict(US_oil_std_df['Oil Production (TWh)'].values.reshape(-1, 1))

### Plot the data
plt.scatter(US_oil_std_df['Oil Production (TWh)'], US_oil_std_df['Temp.Anom'], label = 'Data Points')
plt.plot(US_oil_std_df['Oil Production (TWh)'], temp_preds, color = 'r', label = 'Regression Line')
plt.xlabel('US Oil Production')
plt.ylabel('Temperature Anomalies')
plt.title('US Oil Production vs. Temp Anomalies Linear Regression Model')
plt.legend()
```

Here, we can more easily see the relationship between increased oil production and increased temperature anomalies in the linear regression model.

This model tells us that as oil production increases, so do the temperature anomalies, and vice versa.

Let us now look at a linear regression model for the plastic waste data.

```python
# Plastics Waste
### Create and fit linear regression model to the standardized plastics data
lrm = LinearRegression()
lrm.fit(US_plastics_std_df['Plastic.Waste'].values.reshape(-1, 1), US_plastics_std_df['Temp.Anom'].values)

### Make our prediction (regression) line
temp_preds = lrm.predict(US_plastics_std_df['Plastic.Waste'].values.reshape(-1, 1))

### Plot the data
plt.scatter(US_plastics_std_df['Plastic.Waste'], US_plastics_std_df['Temp.Anom'], label = 'Data Points')
plt.plot(US_plastics_std_df['Plastic.Waste'], temp_preds, color = 'r', label = 'Regression Line')
plt.xlabel('US Plastics Waste')
plt.ylabel('Temperature Anomalies')
plt.title('US Plastics Waste vs. Temp Anomalies Linear Regression Model')
plt.legend()
```

Here in the linear regression model, we can see the relationship much more clearly than in the standard line plot.

This model shows that there is a clear linear relationship between plastic waste and temperature anomalies. Thus, as plastic waste increases, so do temperature anomalies, and vice versa.

### Conclusions

There seems to be a clear positive linear relationship between U.S. Oil Production and the temperature anomalies observed by NOAA in the latitudinal range, 20 degrees North to 90 degrees North.

There also seems to be a clear positive linear relationship between U.S. Plastic Waste and the temperature anomalies observed by NOAA in the latitudinal range, 20 degrees North to 90 degrees North.
